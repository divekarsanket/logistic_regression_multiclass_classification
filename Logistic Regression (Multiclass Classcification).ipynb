# Create and train logistic regression model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

```python
y_test#this is real output 
```

```python
model.predict(X_test)#this output gives model
```

# Measure accuracy of our model

```python
model.score(X_test, y_test) #to check where it fails ie. 4% we use confusion matrix
```

# Confusion Matrix

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_predicted)#computes the confusion matrix for evaluating the performance of a classification model
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
